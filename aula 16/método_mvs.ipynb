```julia
# Variância
σ² = 15
# Média
μ = 5

# Gero uma população com distribuição normal padrão, com o tamanho aproximado da minha cidade.
# Por padrão randn gera números com média 0 e variância 1. Por isso multiplico pelo desvio padrão e somo com a média.
população = randn(321000).*sqrt(σ²).+μ;
```

```julia
s = tamanho_amostra = 10

amostra = rand(população, s);
```

```julia
hat_σ² = (vectorX, μ) -> begin
    sum((vectorX .- μ).^2) / length(vectorX)
end

hat_μ = (vectorX) -> begin
    sum(vectorX) / length(vectorX)
end
```

```julia
μ_estimada = hat_μ(amostra)
σ²_estimada = hat_σ²(amostra, μ_estimada) 
```

```julia
println("μ e σ² estimados com uma amostra de $s são respectivamente $μ_estimada e $σ²_estimada")
println("Os valores reais são: $μ e $σ²") 
```

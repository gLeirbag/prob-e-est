```julia
using Statistics
população = rand(321000);
pop_μ = Statistics.mean(população)
pop_σ² = Statistics.varm(população, pop_μ);
print("Média e Variância: $pop_μ e $pop_σ²")
```

```julia
n = tamamostra = 50000
N = numamostras = 10000
amostras = [rand(população, n) for _ in 1:N]

# Calculando média e variância para cada amostra
amostras_μ = [Statistics.mean(a) for a in amostras]
amostras_σ² = [Statistics.varm(a, m) for (a,m) in zip(amostras, amostras_μ)];
```

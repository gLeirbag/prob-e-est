```python
import numpy
```

```python
dados = [1, 2, 3, 4, 5, 6]
qtd_sort_por_cada_face = numpy.zeros(6, numpy.uint)
n = 5000

generator = numpy.random.Generator(bit_generator= numpy.random.PCG64())
```

```python
for experimento in range(n):
    face_cima = generator.choice(dados)
    qtd_sort_por_cada_face[face_cima - 1] = qtd_sort_por_cada_face[face_cima - 1] + 1

print(qtd_sort_por_cada_face/n)

# print(numpy.average(qtd_sort_por_cada_face))
```

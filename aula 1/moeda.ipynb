```python
import numpy
import matplotlib.pyplot as mplt
```

```python
ea_lançamento_moeda = {'cara', 'coroa'}

n_lançamentos = 1000

experimentos = list()
for lanc_exper in range(1, n_lançamentos):
    resultado = list()
    for i in range(lanc_exper):
        # Vamos criar um gerador de números pseudo-aleatório usando o gerador PCG64, vamos usar o método .choice para
        # selecionar um elemento de um array.
        resultado.append(numpy.random.Generator(bit_generator= numpy.random.PCG64()).choice(list(ea_lançamento_moeda)))
    experimentos.append({
            "num_lanc": lanc_exper,
            "resultado": resultado
        })
        
```

```python
# lista contendo a probabilidade frequentista de cada um dos experimentos (Um experimento faz um lançamento a mais que o experimento anterior)
l_cara = [experimento['resultado'].count('cara')/experimento['num_lanc'] for experimento in experimentos]
mplt.plot(range(1, n_lançamentos), l_cara)

```

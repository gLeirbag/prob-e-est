# Regressão Linear e o Métodos dos Mínimos Quadrados

> O Método dos Mínimos Quadrados é um método para achar a curva que melhor descreve um conjunto de dados através da minimização da soma dos resíduos ao quadrado.

<center><img src="../assets/mmq_scatter_plot.png"/></center>

Tendo que o chamado *resíduo* é definido como
$$\epsilon_{x_i} = y_{x_i} - \hat{y}_{x_i}$$

Quero encontrar a reta $y = ax+b$ que minimize  a soma dos quadrados dos *resíduos*.

$$\argmin_{\hat{y}_{x_i}}{\sum_i^n (y_{x_i} - \hat{y}_{x_i})^2 } \Rightarrow \argmin_{a, b} \sum_i^n (y_{x_i} - ax_i - b)^2 $$

Note pela otimização ser em relação a uma função dependente de 2 parâmetros, devemos encontrar a função gradiente e iguala-lá a 0.

$$
∇ \sum_i^n \epsilon_{x_i}^2 \implies \sum_i^n∇\epsilon_{x_i}^2 = \Biggl[ \sum_i^n \frac {δϵ_{x_i}^2} {δa} , \sum_i^n \frac {δϵ_{x_i}^2} {δb} \Biggr] = 0
$$

Primeiro, vamos começar derivando em relação a $a$.
$$\begin{align*}
    \frac {δϵ_{x_i}^2} {δa} &= \frac {δ} {δa} (y_{x_i} - ax_i - b)^2 \\
    &\xlongequal{\text{cadeia}} \frac {δu^2} {δu} * \frac {δ} {δa} (y_{x_i} - ax_i - b)^2, \text{ com }u = y_{x_i} - ax_i - b \\
    &= 2u * (0 - x_i + 0) = 2u(-x_i) = -2x_iu \implies 2x_i(ax_i + b - y_{x_i}) \\
    ∴ \frac {δϵ_{x_i}^2} {δa} &= 2x_i(ax_i + b - y_{x_i})
\end{align*}$$

Agora, em relação a $b$.

$$\begin{align*}
    \frac {δϵ_{x_i}^2} {δb} = ... &= \frac {δu^2} {δu} * \frac {δ} {δb} (y_{x_i} - ax_i - b)^2 \\
    &= -2u \implies 2(ax_i+b-y_{x_i}) \\
    ∴ \frac {δϵ_{x_i}^2} {δb} &= 2(ax_i + b - y_{x_i})
\end{align*}$$

Logo 
$$ \sum_i^n∇\epsilon_{x_i}^2(a,b) = \biggl[ 2 \sum_i^n x_i(ax_i+b-y_{x_i}) , 2\sum_i^n (ax_i+b-y_{x_i})\biggr] = 0 \\

∇f(a,b) = \biggl[ 2 \sum_i^n x_i(ax_i+b-y_{x_i}) , 2\sum_i^n (ax_i+b-y_{x_i})\biggr] = 0 

$$

Igualando ambas as derivadas a 0, teremos:
$$
\def\xsum{\sum_i^nx_i}

\begin{gather}
a\xsum^2+ b\xsum - \xsum y_{x_i} &= 0 \\
a\xsum+ nb - \sum_i^n y_{x_i} &= 0
\end{gather}
$$
Da eq. 2, conseguimos:
$$
\def\xsum{\sum_i^nx_i}

b = \frac {\sum_i^n y_{x_i} - a\xsum} {n} \tag {3}

$$


Da 1, usando a 3:

$$
\def\xsum{\sum_i^nx_i}
\begin{align*}
\def\xsum{\sum_i^nx_i}

&a\xsum^2+ \frac {\sum_i^n y_{x_i} - a\xsum} {n} * \xsum - \xsum y_{x_i} = 0 \\ 
\implies &
a\xsum^2+ \frac {\sum_i^n y_{x_i} \xsum - a(\xsum)^2} {n} - \xsum y_{x_i} = na\xsum^2+ \sum_i^n y_{x_i} \xsum - a\Bigl(\xsum\Bigr)^2 - n \xsum y_{x_i} = 0 \\
\implies &
a\Bigg(n \xsum^2 - \Big(\xsum\Big)^2\Bigg) = n \xsum y_{x_i} - \sum_i^n y_{x_i} \xsum \\
\end{align*}

\\
\scriptsize\text{ Lembrar que a soma dos quadrados não é igual ao quadrado das somas!}
$$
$$
∴ a = \frac {n\xsum y_{x_i} - \sum_i^n y_{x_i} \xsum}{n \xsum^2 - \big(\xsum\big)^2} \tag{4}
$$

Para termos $b$ sem ser em função de $a$, substituímos 4 em 1. _Se substituíssemos 4 em 3, teríamos de volta a equação 1, o que não seria útil_.

$$
\def\xsum{\sum_i^nx_i}
∴ b = \frac {\sum_i^n y_{x_y}\xsum² - \xsum y_{x_i} \xsum}{n \xsum^2 - \big(\xsum\big)^2}$$

Finalmente! Agora tudo que precisamos para calcular a reta que melhor descreve o conjunto de pontos é: a soma de x, a soma de y, a soma de x², a soma de x*y.

<center><img src="../assets/mmq_scatter_plot_2.png"/></center>


Ainda podemos manipular a equação para usarmos termos ainda mais estatísticos:

$$
\def\xsum{\sum_i^nx_i}
\begin{gather*}
    a &= \frac {(n\xsum y_{x_i} - \sum_i^n y_{x_i} \xsum ) * \frac{1}{n^2}}{ ( n \xsum^2 - \big(\xsum\big)^2 ) * \frac{1}{n^2}} \\\\
    a &= \frac { \bar{xy} - \bar{y}\bar{x} }{ \bar{x^2}  - \bar{x}^2 } \implies a = \frac{\text{cov}(x,y)}{\text{var}(x)} \\\\
    & \therefore \hat{a} = \frac {\text{cov}(x,y)} {\text{var}(x)} \\\\
\end{gather*} \\
\begin{gather*}
    e\ \ 
    \hat{b} = \bar{y} - a\bar{x}
\end{gather*}
$$

## Coeficiente de Determinação($R^2$)

O *coeficiente de determinação*, R², é um escalar que representa a qualidade do quanto um conjunto de dados observados se adéquam a um modelo de predição.

No caso de regressão linear, ele é dador por:
$$
R^2 = 1 -\frac{\text{Soma dos Quadrados dos Resíduos}}{\text{Soma Total dos Quadrados*}}= \frac {\sum(y_i-\hat{y_i})^2} {\sum(y_i-\bar{y})^2}

\\\\
\footnotesize\text{*A Soma Total dos Quadrados é o escalar dado pela soma do quadrado da diferença de cada um dos dados pela sua média.} //
$$

O coeficiente de determinação ainda pode ser escrito como a *correlação de pearson* ao quadrado.

$$
R^2 = \frac {\text{cov}^2 (x,y) } {S_{xx}S_{yy}}
$$


```julia
quantidadedados = 300
coeangular = 4
coelinear = 3
fx = (x) -> x*coeangular+coelinear
yobservado = zeros(quantidadedados)

for i in 1:quantidadedados
   yobservado[i] = fx(i) + randn()*abs(quantidadedados/2-i)
end
```

```julia
using CairoMakie
figtitle = "Variável X e Y"
fig = Figure(size = (400, 400), backgroundcolor = :transparent)
axis = Axis(fig[1,1], title = figtitle, titlesize = 18, xlabel = "X", ylabel = "Y", backgroundcolor = :white)

Makie.inline!(true)

scatter!(axis, 1:300, yobservado, color = :tomato)
save("../assets/mmq_scatter_plot.png", fig)
```

```julia
somas = ( x = sum(1:quantidadedados), y = sum(yobservado), xy = sum((1:quantidadedados).*yobservado), xx = sum((1:quantidadedados) .^2) ) ;
```

```julia
using Statistics

function calcular_pcoefangular(somas, n)
    (n * somas.xy - somas.x * somas.y) / (n * somas.xx - somas.x^2)
end

function calcular_pcoeflinear(somas, n, coelinear)
    (somas.y - coelinear * somas.x) / n
end

function calcular_coedeterminação(amostra, (coeangular, coelinear))
    STQ = sum( (amostra .- mean(amostra)) .^ 2 )
    SRQ = sum( (amostra .- (coeangular .* (1:length(amostra))) .+ coelinear) .^ 2 )
    1 - (SRQ / STQ)
end
```

```julia
pcoefangularestimado = calcular_pcoefangular(somas, quantidadedados)
pcoeflinearestimado = calcular_pcoeflinear(somas, quantidadedados, pcoefangularestimado)
coedeterminação = calcular_coedeterminação(yobservado, (pcoefangularestimado, pcoeflinearestimado));
```

```julia
ablines!(axis, [pcoeflinearestimado, coelinear], [pcoefangularestimado, coeangular]; color = [:orange, :blue], label = ["MMQ", "Reta Original"]);
text!(axis, "R² = $(round(coedeterminação, digits=4))", position = (200, 120), color = :black, align = (:left, :bottom), fontsize = 12)

axislegend(axis, position = :rb)
save("../assets/mmq_scatter_plot_2.png", fig)
```

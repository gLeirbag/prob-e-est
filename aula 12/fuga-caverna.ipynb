# Problema do Trabalhador Perdido na Mina

Um mineiro se perde dentro de uma mina, diante dele há 3 caminhos diferentes. O caminho A pode ser percorrido por 2 horas e o levará até a saída. O caminho B tem um período de caminhada de 3 horas e leva até o ponto inicial. O caminho C também o faz retornar ao ponto de origem, mas, em 5 horas. Qual o tempo esperado que o trabalhador ficará dentro da mina (suponde que ele sempre esqueça qual caminho percorreu).

```julia
paths = [2, 3, 5]
```

```julia
function try_to_exit()
    total_hours = 0
    exit_found = false
    while !exit_found
        chosen = rand(paths)
        total_hours += chosen
        if chosen == 2
            exit_found = true
        end
    end
    total_hours
end
```

```julia
function repeat_experiments(n)
    [try_to_exit() for n in 1:n]
end
```

```julia
e_x_n = expectancy_by_number_of_experiments = [ (n = n, mean = Statistics.mean(repeat_experiments(n))) for n in 1:5000 ]
```

```julia
ns = get.(e_x_n, :n, Nothing)
means = get.(e_x_n, :mean, Nothing)
```
